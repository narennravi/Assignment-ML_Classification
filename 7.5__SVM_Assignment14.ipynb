**Import libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
     
```

```python
from sklearn.svm import SVC
```

#-----------------------------------------------------------------------------

**Importing the dataset**

```python
df = pd.read_csv('/content/7.3__Social_Network_Ads.csv')
```

#-----------------------------------------------------------------------------

**Splitting the dataset into the Training set and Test set**

```python
X = df[['Age','EstimatedSalary']] # Only these two features have been taken for modelling as per the instructions in the assignment pblm statement
Y = df['Purchased']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size=0.2, random_state=5, stratify=Y)
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

#-----------------------------------------------------------------------------

**Feature Scaling**

```python
from sklearn.preprocessing import StandardScaler
func = StandardScaler()
func.fit(x_train)
x_train_std = func.transform(x_train)
x_test_std = func.transform(x_test)
```

#-----------------------------------------------------------------------------

**Fitting SVM to the Training set**

```python
clf = SVC(kernel="linear")
clf.fit(x_train_std,y_train)
clf.score(x_test_std,y_test)
```

#-----------------------------------------------------------------------------

**Predicting the Test set results**

```python
y_pred = clf.predict(x_test_std)
```

#-----------------------------------------------------------------------------

**Making the Confusion Matrix**

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred) 
```

```python
sns.heatmap(cm, annot=True)
```

#-----------------------------------------------------------------------------

**Visualising the Training set results**

```python
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

xs = x_train_std[:,0]
ys = x_train_std[:,1]
zs = y_train
ax.scatter(xs, ys, zs, s=50,color=['purple' if p==1 else 'cyan' for p in clf.predict(x_train_std)], alpha=0.6, edgecolors='w')

ax.set_xlabel('Age')
ax.set_ylabel('EstimatedSalary')
ax.set_zlabel('Purchased')
```

#-----------------------------------------------------------------------------

**Visualising the Test set results**

```python
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

xs = x_test_std[:,0]
ys = x_test_std[:,1]
zs = y_test
ax.scatter(xs, ys, zs, s=50,color=['purple' if p==1 else 'cyan' for p in y_pred], alpha=0.6, edgecolors='w')

ax.set_xlabel('Age')
ax.set_ylabel('EstimatedSalary')
ax.set_zlabel('Purchased')
```
